## What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.

**filtrar las peliculas que fueron lanzadas en 1990**

```python
#almacenamos en movies_90s el dataframe y seleccionamos el tipo de pelicula y el rango de años que nos interesan.
netflix_subset = netflix_df[(netflix_df["type"] == "Movie")] 
subset = netflix_subset[(netflix_subset["release_year"] >= 1990)]
movies_90s = subset[(subset["release_year"] < 2000)]
```

```python
plt.hist(movies_90s["duration"], bins=20)
plt.xlabel('Duration')
plt.ylabel('Count')
plt.title('Distribution of Movie Durations in the 90s')
plt.show()
duration = 100
```

## A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

**Filtrar por peliculas de accion**

```python
action_movies_90s = movies_90s[(movies_90s['genre'].str.contains('Action'))]
```

**Filtrar las peliculas de menos de 90 min**

```python
short_action_movies_90s = action_movies_90s[action_movies_90s ['duration'] < 90]
```

**Contamos la cantidad de peliculas menores a 90 min**

```python
short_movie_count = len(short_action_movies_90s)
```

**Mostrar resultado**

```python
print(short_movie_count)
```
